**6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
# Define a function to calculate the median value of a given dataset 'd'
def medain_value(d):
  # Sort the data in ascending order
  sorted_data=sorted(d)
  n = len(sorted_data)

  if n % 2 == 0:
     # If the number of data points is even, calculate the median as the average of the two middle values
    med = (sorted_data[n//2 - 1] + sorted_data[n//2]) // 2

  else:
    # If the number of data points is odd, the median is the middle value
    med = sorted_data[n//2]
  return  med

# Calculate the median value of the column
median=medain_value(df['housing_median_age'])
# Print the calculated median value
print(median)
```

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
# Define a function to categorize the size of 'total_bedroom' based on conditions
def newclom(total_bedroom):
  if total_bedroom <= 10:
    return 'Small'
  elif 10 < total_bedroom < 1000:
    return 'Medium'
  else:
    return 'large'
# Apply the 'newclom' function to 'total_bedrooms' column of the DataFrame 'df' and create new column 'total_bedroom_size'
df['total_bedroom_size'] = df['total_bedrooms'].apply(newclom)
# Display the first few rows of the DataFrame 'df' with the newly added 'total_bedroom_size' column
df
```
