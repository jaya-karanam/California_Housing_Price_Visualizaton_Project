# **CALIFORNIA HOUSING DATASET**

## **INTRODUCTION**

**FEATURE EXPLANATION:**

The California Housing Dataset contains 10 columns which contains data drawn from the 1990 U.S. Census. Here is the discription of each column:

1. **longitude**: A measure of how far west a house is; a more negative value is farther west.
2. **latitude**: A measure of how far north a house is; a higher value is farther north.
3. **housing_median_age**: Median age of a house within a block; a lower number is a newer building.
4. **total_rooms**: Total number of rooms within a block.
5. **total_bedrooms**: Total number of bedrooms within a block.
6. **population**: Total number of people residing within a block.
7. **households**: Total number of households, a group of people residing within a home unit, for a block.
8. **median_income**: Median income for households within a block of houses (measured in tens of thousands of US Dollars).
9. **median_house_value**: Median house value for households within a block (measured in US Dollars).
10. **ocean_proximity**: Distance from the house to the ocean/sea.



**FEATURE TYPES:**

1.   **Discrete data**: housing_median_age, total_rooms, total_bedrooms, population, households, median_house_value.
2.   **Continuous data**: longitude, latitude, median_income.
3.   **Nominal data**: ocean_proximity.

### **Q6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
# Creating function called cal_median_value to find median value of the data
def cal_median_value(values):
  median = values.median()        # Finding median value
  return median                   # Returning the median value
```

### **Q10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
# Making a copy of nonull_housing and creating a new dataframe called cleaned_housing
cleaned_housing = nonull_housing.copy()
cleaned_housing
```

```python
# Converting total_bedrooms column into integer type
cleaned_housing['total_bedrooms'] = cleaned_housing['total_bedrooms'].astype(int)
cleaned_housing
```

```python
# Creating new column named total_bedroom_size in cleaned_housing dataframe using the lambda function based on condition in the question
cleaned_housing['total_bedroom_size'] = cleaned_housing['total_bedrooms'].apply(lambda tb : 'small' if tb <= 10 else ('medium' if tb >= 11 and tb < 1000 else 'large'))
```

```python
# Displaying cleaned_housing dataframe
cleaned_housing
```

```python
# Displaying the counts of unique value in the total_bedroom_size
cleaned_housing['total_bedroom_size'].value_counts()
```

**Explanation:** There are 55 small, 18495 medium and 1883 large bedroom size are there in blocks.
